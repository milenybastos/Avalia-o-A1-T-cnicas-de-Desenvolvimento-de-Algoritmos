```python
# --- Tarefa 4: Cadastro de Produtos (Dicionários) ---

print("==============================================")
print("     Cadastro de Produtos (usando Dict)     ")
print("==============================================")
print("Digite 'fim' no nome do produto para parar.")

# 1. Inicialização do Dicionário
# Podemos usar dict() ou, mais comum, {}
produtos = {}



# 2. Entrada de Dados
while True:
    nome = input("\nDigite o nome do produto: ")

    # Condição de parada
    if nome.lower() == 'fim':
        break

    # Se o produto já existe, avisamos e pulamos para o próximo
    if nome in produtos:
        print(f"Produto '{nome}' já cadastrado. Use outro nome.")
        continue # Pula o resto do código e volta pro início do 'while'

    # 3. Inserção de Dados (com tratamento de erro para o preço)
    try:
        # Pedimos o preço e tentamos converter para float (número com decimal)
        preco = float(input(f"Digite o preço para '{nome}': R$ "))

        # Verificação para não aceitar preços negativos
        if preco < 0:
            print("Preço inválido. Digite um valor positivo.")
            continue # Volta pro início do 'while'

        # --- PONTO PRINCIPAL ---
        # Inserção no dicionário:
        # A 'chave' (key) é o 'nome', e o 'valor' (value) é o 'preco'.
        produtos[nome] = preco
        print(f"Produto '{nome}' - R$ {preco:.2f} cadastrado!")

    except ValueError:
        print("Erro: O preço deve ser um número (ex: 10.50). Produto não cadastrado.")

print("\n==============================================")

# 4. Recuperação e Impressão dos Dados
if len(produtos) == 0:
    print("Nenhum produto foi cadastrado.")
else:
    print(f"--- Lista Final de Produtos ({len(produtos)}) ---")

    # Usamos .items() para pegar a chave (nome) E o valor (preco)
    # em cada "volta" do laço.
    for nome, preco in produtos.items():
        # :.2f formata o preço para sempre ter 2 casas decimais
        print(f"Produto: {nome} | Preço: R$ {preco:.2f}")

print("==============================================")
```
