```python
# --- Tarefa 2: Contador de Números Pares (1 a 100) ---

print("==============================================")
print("     Contador de Pares com 'for' e 'while'    ")
print("==============================================")


# --- Versão com 'for' ---
# O 'for' é ótimo quando você sabe o começo e o fim.
# O range(1, 101) gera números de 1 até 100. (A pegadinha é que o 101 fica de fora!)
print("\n--- Números Pares (usando FOR) ---")

for numero in range(1, 101):

    # A lógica para checar se é par:
    # Usamos o operador módulo (%) para pegar o RESTO da divisão por 2.
    # Se o resto for 0, o número é par.
    if numero % 2 == 0:
        print(numero)


print("\n==============================================")
print("Fim da contagem!")
```

```python


# --- Versão com 'while' ---
# O 'while' exige mais controle manual. Você precisa de 3 coisas:
# 1. Onde começa (contador = 1)
# 2. Quando para (while contador <= 100)
# 3. Como avança (contador += 1)
print("\n--- Números Pares (usando WHILE) ---")

contador = 1 # 1. Onde começa

while contador <= 100: # 2. Quando para

    # A lógica de checagem é exatamente a mesma
    if contador % 2 == 0:
        print(contador)

    # 3. Como avança (O INCREMENTO)
    # ESSENCIAL! Se esquecer esta linha, o 'contador' nunca muda
    # e o programa entra em loop infinito.
    contador = contador + 1 # ou a forma curta: contador += 1

print("\n==============================================")
print("Fim da contagem!")
```
