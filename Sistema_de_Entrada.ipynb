```python


print("==============================================")
print("     Sistema de Entrada   ")
print("==============================================")

# 1. Tentar pegar a idade e já converter
try:
    # input() sempre devolve uma string (texto)
    entrada = input("Qual a sua idade? ")

    # int() tenta converter a string para um número inteiro
    # Se o usuário digitar "abc", isso vai falhar e pular pro 'except'
    idade = int(entrada)

# 2. O "Guarda" (Guard Clause)
    # Primeiro, vamos barrar as idades sem sentido (negativas ou muito altas).
    # Se a idade for inválida, a gente já para o código aqui.
    if idade <= 0 or idade > 120:
        print(f"\n[IDADE INEXISTENTE] ")
        print(f"Idade '{idade}'? Isso não é válido. Tente de novo.")

# 3. Lógica Principal (O "Caminho Feliz")
    # Se o código chegou até aqui, a idade é um número válido (ex: 15, 20, 50).

    # Regra 1: Maior de idade
    elif idade >= 18:
        print(f"\n[APROVADO] ")
        print(f"Idade: {idade}. Acesso total. Pode pegar a pulseira VIP.")

    # Regra 2: 16 ou 17 anos
    # Só entra aqui se o 'if' de cima (>= 18) for falso.
    elif idade >= 16:
        print(f"\n[APROVADO COM RESTRIÇÃO] ")
        print(f"Idade: {idade}. Acesso à pista, mas longe do bar. Pulseira Jovem.")

    # Regra 3: Menor de 16
    # Se nenhum dos 'if' ou 'elif' acima foi verdadeiro, cai aqui.
    else:
        print(f"\n[BARRADO] ")
        print(f"Idade: {idade}. Desculpe, festa só para maiores de 16.")

# 4. Tratamento de Erro (O Famoso "try...except")
# Se o 'try' lá em cima falhar (ex: digitar "vinte"), o Python pula direto pra cá.
except ValueError:
    print(f"\n[ERRO DE ENTRADA] ")
    print(f"O que você digitou ('{entrada}') não é um NÚMERO.")
    print("Rode o programa de novo e digite sua idade com números.")

print("\n==============================================")
```
