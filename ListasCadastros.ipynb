```python
# --- Tarefa 3: Cadastro de Alunos em Lista ---

print("==============================================")
print("     Cadastro de Alunos (usando Listas)     ")
print("==============================================")
print("Digite os nomes dos alunos. Digite 'fim' para parar.")

# 1. Inicialização da Lista
# Começamos com uma lista vazia que vai "segurar" os nomes.
nomes_alunos = []

# 2. Entrada Dinâmica de Dados (Laço)
while True:
    # Pede um nome ao usuário em cada repetição do laço
    nome = input("Digite o nome do aluno: ")

    # 3. Verificação de Parada
    # Usamos .lower() para garantir que "fim", "Fim" ou "FIM" funcionem.
    if nome.lower() == 'fim':
        break # Sai imediatamente do laço 'while'

    # Se o nome não for "fim", adicionamos ele à lista
    nomes_alunos.append(nome)
    print(f"'{nome}' adicionado. (Total: {len(nomes_alunos)})") # Feedback legal pro usuário

print("\n==============================================")

# 4. Impressão Correta dos Elementos
# Verificamos se a lista não está vazia antes de tentar imprimir
if len(nomes_alunos) == 0:
    print("Nenhum aluno foi cadastrado.")
else:
    print(f"--- Lista Final de Alunos Cadastrados ({len(nomes_alunos)}) ---")

    # Usamos um 'for' para percorrer cada item da lista e imprimir
    # 'aluno' é uma variável temporária que assume o valor de cada item
    # a cada "volta" do laço.
    for aluno in nomes_alunos:
        print(f"- {aluno}")

print("==============================================")
```
